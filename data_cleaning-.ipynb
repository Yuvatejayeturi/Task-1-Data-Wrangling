```python
import pandas as pd

df = pd.read_csv("netflix_titles.csv")

print("Shape:", df.shape)
print("\nColumns:")
print(df.columns)

print("\nDataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum())

print("\nDuplicates:")
print(df.duplicated().sum())
```
